```python
from ASDM.ASDM.Engine import Structure
import pandas as pd
pd.options.display.max_columns = 100
```

```python
# Compile the model from XMILE

model = Structure(from_xmile='Elective Recovery Model.stmx')
```

```python
model.clear_last_run()
model.simulate()
```

```python
df_outcome = model.export_simulation_result()
```

```python
df_outcome['"13wk wait for urgent treatment"'].plot()
```

### Prepare ASDM simulation DataFrame

```python
# print(df_outcome.columns)

asdm_outcome = df_outcome[sorted(df_outcome.columns)]
asdm_outcome
```
